```python
# Lab Assignment 1

source_file = input("Enter source file name: ")
destination_file = input("Enter destination file name: ")

# Read from source file
with open(source_file, 'r') as f:
    content = f.read()

# Convert to uppercase
upper_content = content.upper()

# Write to destination file
with open(destination_file, 'w') as f:
    f.write(upper_content)

print("Content copied in uppercase successfully!")

# Display content
print("\n--- Source File Content ---")
print(content)

print("\n--- Destination File Content ---")
print(upper_content)
```

```python
# Lab Assignment 2

source_file = input("Enter source Python file name: ")
destination_file = input("Enter destination file name: ")

with open(source_file, 'r') as src, open(destination_file, 'w') as dest:
    for line in src:
        # Remove comments
        stripped_line = line.split('#')[0]

        # Write only non-empty lines
        if stripped_line.strip() != "":
            dest.write(stripped_line + '\n')

print("File copied without comments successfully!")

# Display both files
print("\n--- Source File Content ---")
with open(source_file, 'r') as f:
    print(f.read())

print("\n--- Destination File Content ---")
with open(destination_file, 'r') as f:
    print(f.read())
```
