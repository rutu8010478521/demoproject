```python
import matplotlib.pyplot as plt
sales=[210,270,300,350,400,300,200,600]
year=[2011,2012,2013,2014,2015,2016,2017,2018]
plt.plot(year,sales)
plt.title("sales of johnson company")
plt.xlabel("year")
plt.ylabel("sales")
plt.show()
```

```python
import matplotlib.pyplot as plt
pen=[210,270,300,350,400,300,200,600]
copy=[300,320,360,400,200,120,300,350]
year=[2011,2012,2013,2014,2015,2016,2017,2018]
plt.plot(year,pen,color="black",label="pen",linestyle="-",linewidth=2,marker="o",markersize=5)
plt.plot(year,copy,color="pink",label="copy",linestyle=":",linewidth=4,marker="o",markersize=5)
plt.title("sales of johnson company")
plt.xlabel("year")
plt.ylabel("sales")
plt.legend()  #shows coner  points x` is and y is
plt.show()
```

```python
study_hours = [1, 2, 3, 4, 5]
marks = [20, 40, 60, 80, 100]

# Create the plot
plt.plot(study_hours, marks, label="Marks and hours",
         linestyle='-', color='blue', marker='o')

# Add title and labels
plt.title("Study Hours vs Marks")
plt.xlabel("No. of hours")
plt.ylabel("Marks")

# Add legend
plt.legend()

# Display the plot
plt.grid(True) # adds a grid to the graph.its shows grid
plt.show()
```

```python
import matplotlib.pyplot as plt
# a) Accept a single integer N (number of data points)
n = int(input("Enter the number of data points (N): "))
 # Lists to store the axis data
study_hours = []
marks_obtained = []

# b) Accept N integer values for the X-axis (Study Hours)print(f"\nEnter {n} values for Study Hours (X-axis):")
for i in range(n):
   val = int(input(f"  Value {i+1}: "))
   study_hours.append(val)

 # c) Accept N integer values for the Y-axis (Marks Obtained)
print(f"\nEnter {n} values for Marks Obtained (Y-axis):")
for i in range(n):
  val = int(input(f"  Value {i+1}: "))
  marks_obtained.append(val)

  # d) Generate the line plot with specific customizations
  # color='blue', linestyle='-' (solid), marker='s' (square)
plt.plot(study_hours, marks_obtained, color='blue',
                 linestyle='-', marker='s', label='Performance Data')
 # Set Title
plt.title("Student Performance Analysis")

# Set Axis Labels
plt.xlabel("Hours Studied")
plt.ylabel("Marks (0-100)")

 # Add Legend
plt.legend()

  # Display the plot
plt.show()
```
