```python
# Lab Assignment 1

# Input: Ask the user to enter a string
input_string = input("Enter a string: ")

# a) Number of Vowels
vowels = "AEIOUaeiou"
vowel_count = sum(1 for char in input_string if char in vowels)
print(f"Number of Vowels: {vowel_count}")

# b) Number of Consonants
consonant_count = sum(1 for char in input_string if char.isalpha() and char not in vowels)
print(f"Number of Consonants: {consonant_count}")

# c) Number of Spaces
space_count = input_string.count(' ')
print(f"Number of Spaces: {space_count}")

# d) Number of Lowercase Letters
lowercase_count = sum(1 for char in input_string if char.islower())
print(f"Number of Lowercase Letters: {lowercase_count}")
```

```python
# Lab Assignment 2

# Function to accept sequence of lines, capitalize, and print
def capitalize_lines(input_lines):
    # Capitalizing the entire line and printing
    for line in input_lines:
        print(line.upper())

# Input: Taking multiple lines from the user
print("Enter multiple lines (enter an empty line to stop):")
lines = []
while True:
    line = input()
    if line == "":
        break
    lines.append(line)

# Call the function to capitalize and print the lines
capitalize_lines(lines)
```
