```python
# Online Polling & Voting System

candidates = ["Alice", "Bob", "Charlie"]
votes = [0, 0, 0]   # vote count for each candidate
voters = []         # to prevent duplicate voting

def display_candidates():
    print("\nCandidates:")
    for i in range(len(candidates)):
        print(f"{i+1}. {candidates[i]}")

def vote():
    voter_id = input("Enter your voter ID: ")

    if voter_id in voters:
        print("You have already voted! Duplicate voting not allowed.")
        return

    display_candidates()

    try:
        choice = int(input("Select candidate number to vote: "))

        if 1 <= choice <= len(candidates):
            votes[choice-1] += 1
            voters.append(voter_id)
            print("Vote recorded successfully!")
        else:
            print("Invalid candidate number.")

    except:
        print("Invalid input!")

def show_results():
    print("\nCurrent Poll Results:")
    total_votes = sum(votes)

    for i in range(len(candidates)):
        if total_votes == 0:
            percent = 0
        else:
            percent = (votes[i] / total_votes) * 100

        print(f"{candidates[i]}: {votes[i]} votes ({percent:.2f}%)")

def winner():
    max_votes = max(votes)
    index = votes.index(max_votes)
    print(f"\nCurrent Leader: {candidates[index]} with {max_votes} votes")

def menu():
    while True:
        print("\n--- Online Polling System ---")
        print("1. Cast Vote")
        print("2. Show Poll Statistics")
        print("3. Show Winner")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            vote()
        elif choice == "2":
            show_results()
        elif choice == "3":
            winner()
        elif choice == "4":
            print("Exiting Poll System")
            break
        else:
            print("Invalid choice!")

menu()
```
