```python
nums = [10, 45, 23, 89, 12]

largest = max(nums)
print("Largest number:", largest)
```

```python
nums = [1, 2, 3, 4, 5, 6]

result = [n for n in nums if n % 2 != 0]
print(result)
```

```python
nums = [1, 2, 3, 4, 5]

nums.reverse()
print(nums)
```

```python
nums = [1, 2, 3, 4]

squares = [n**2 for n in nums]
print(squares)
```

```python
num = ["Hello", 34, 45, "", 40]

result = [x for x in num if x != ""]
print(result)
```

```python
nums = [10, 12, 15, 22, 25, 30]

result = [n for n in nums if n % 5 == 0]
print(result)
```

```python
num = [2,3,4,5,2,6,3,2]

result = list(set(num))
result.sort()
print(result)
```

```python
list1 = [23, 24, 25, 26]
list2 = [27, 28, 29, 30]

list1.extend(list2)
print(list1)
```

```python
nums = [2, 23, 24, 51, 46, 67]

even = [n for n in nums if n % 2 == 0]
odd = [n for n in nums if n % 2 != 0]

print("Even:", even)
print("Odd:", odd)
```

```python
nums = [5, 2, 9, 1, 7]

nums.sort()
print(nums)
```

```python
lst = ['a', 'b', 'c', 'd']

rev = []
for i in range(len(lst)-1, -1, -1):
    rev.append(lst[i])

print(rev)
```
