```python
# Lab Assignment 1

# Input: Ask the user to enter integers separated by space
numbers = tuple(map(int, input("Enter a series of integers separated by space: ").split()))

# a) Print the total number of items in the Tuple
print(f"Total number of items in the Tuple: {len(numbers)}")

# b) Print the last item in the Tuple
print(f"Last item in the Tuple: {numbers[-1]}")

# c) Print the Tuple elements in reverse order
print(f"Tuple elements in reverse order: {numbers[::-1]}")

# d) Print Yes if the Tuple contains integer 5, otherwise No
if 5 in numbers:
    print("Yes, the Tuple contains the integer 5.")
else:
    print("No, the Tuple does not contain the integer 5.")

# e) Remove the first and last items, sort the remaining items, and print the result
modified_tuple = numbers[1:-1]
sorted_tuple = sorted(modified_tuple)
print(f"Tuple after removing the first and last items and sorting: {sorted_tuple}")
```

```python
# Lab Assignment 2

# Input: Ask the user to enter the prices of items sold separated by space
prices = tuple(map(float, input("Enter the prices of items sold separated by space: ").split()))

# a) Print the total number of items sold
print(f"Total number of items sold: {len(prices)}")

# b) Print the price of the cheapest item sold
print(f"Price of the cheapest item sold: {min(prices)}")

# c) Print the price of the costliest item sold
print(f"Price of the costliest item sold: {max(prices)}")

# d) Print the item prices in ascending order
sorted_prices = sorted(prices)
print(f"Item prices in ascending order: {sorted_prices}")

# e) Print the number of costliest items sold
costliest_item_count = prices.count(max(prices))
print(f"Number of costliest items sold: {costliest_item_count}")
```
