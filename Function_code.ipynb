```python
#function call without argument without return
def evenodd():
    num=int(input("Enter a number:"))
    if num%2==0:
        print(num,"even number")
    else:
        print(num, "odd number")

evenodd()
```

```python
#function call with argument pass  without return
def evenodd(n):
    if  n% 2 == 0:
        print(n, "even number")
    else:
        print(n, "odd number")

num = int(input("Enter a number:"))
evenodd(num)  #pass argument

```
