```python
s = "Python is very easy language"
words = s.split()

for w in words:
    if len(w) % 2 == 0:
        print(w)
```

```python
s = "Hello World"
count = 0

for ch in s.lower():
    if ch in "aeiou":
        count += 1

print("Vowel count:", count)
```
